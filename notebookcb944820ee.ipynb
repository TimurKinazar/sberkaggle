```python
cards = {'VI':1,
         'MA':2,
         'EX':3, 
         'PC':4, 
         'ST':5,
         'UY':6
             }
```

```python
from collections import defaultdict
defdict = defaultdict(lambda: [])
for i in range(len(traain)):
    temp = traain.iloc[i]
    defdict[temp['image']].append([(temp['x1'] + temp['w'] / 2)/temp['width'], (temp['y1'] + temp['h'] / 2)/temp['height'], temp['w']/temp['width'], temp['h']/temp['height'], temp['label']])
```

```python
from pathlib import Path
def create_dataset(dictr):
#   images_path = Path(f"clothing/images/{dataset_type}")
#   images_path.mkdir(parents=True, exist_ok=True)
    labels_path = Path("./yolov5/data/labels/train/")
    labels_path.mkdir(parents=True, exist_ok=True)
    for img_id, row in enumerate(tqdm(dictr)):
        image_name = row
        # img = urllib.request.urlopen(row["content"])
        # img = Image.open(img)
        # img = img.convert("RGB")
        # img.save(str(images_path / image_name), "JPEG")
        label_name = f"{image_name[:-4]}.txt"
        with (labels_path / label_name).open(mode="w") as label_file:
            for a in dictr[row]:
                category_idx = cards[a[4]] - 1
            # points = a['points']
            # p1, p2 = points
            # x1, y1 = a[0], 
            # x2, y2 = p2['x'], p2['y']
            # bbox_width = x2 - x1
            # bbox_height = y2 - y1
                label_file.write(
                    f"{category_idx} {a[0]} {a[1]} {a[2]} {a[3]}\n"
                )
```

```python
import os
# if pathofweights.exists():
#     paths = "runs/detect/exp2/labels/"
# else:
paths = "runs/detect/exp2/labels/"
listdirlabels = os.listdir(paths)
namess = []
bb = []
# for i in listdirlabels:
for i in listdirlabels:
    namess.append(i[:-4] + '.jpg')
    f = open(paths+i, 'r')
    tbb = []
    for line in f:
        a = [float(j) for j in line.split()]
        tbb.append(a[1:]+[cardss[a[0]]])
    bb.append(tbb)
tempnotfound = list(set(sub['image']) - set(namess))
tempnbb = [[]]*len(tempnotfound)
namess = namess + tempnotfound
bb = bb + tempnbb
def c_sort(sub_li):
    sub_li.sort(key = lambda x: (x[0], x[1]))
    return sub_li
for i in range(len(namess)):
    bb[i] = c_sort(bb[i])
labels = []
for i in range(len(bb)):
    labb = []
    for j in range(len(bb[i])):
        # xs = namess[i][j][:4]
        labb.append(bb[i][j][4])
    labels.append(" ".join(labb))
dictionar = dict(zip(namess, labels))
labels2 = []
for i in sub['image']:
    labels2.append(dictionar[i])
pd.DataFrame({'image': sub['image'], 'payment': labels2}).to_csv('submission.csv', index=None)
```
